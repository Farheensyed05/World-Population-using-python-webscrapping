### 1)Please scrape the table properly using BeautifulSoup.

```python
import requests # It makes requests to a web page  
from bs4 import BeautifulSoup as bs # It is Python library that is used for web scraping purposes to pull the data out of HTML file
import pandas as pd # It is used for Data Cleaning
import numpy as np #It is used for Mathematical Operations
import matplotlib.pyplot as plt#It is used for Data Visualization
import seaborn as sns#It is used for Data Visualization
```

```python
url="https://www.worldometers.info/world-population/population-by-country/"
source=requests.get(url)#It gets the web page data
```

```python
soup=bs(source.text,"html.parser")#Using Beautiul Soup we can get html data all tags
print(soup.prettify())#It prints the data in a structured manner
```

```python
table=soup.find('table')#Here it finds a table of the html 
print(table.prettify())
```

```python
rows=soup.find('tr')#Here it finds a table row 
print(rows.prettify())
```

```python
data=soup.find_all('td')#with the help of find all method here all table data is printed
print(data)
```

```python
rows=soup.find_all('tr')
rows_list=list()#empty list for all table data to get appended
for tr in rows[1:]:
    td=tr.find_all('td')#getting all data from the table row of table data
    row=[i.text for i in td]# and it gets appended into the row list
    rows_list.append(row)#all table data is appended into the  empty row list
print(rows_list)
```

### 2) Please rename the columns appropriately

```python
df=pd.DataFrame(rows_list,columns=['Ranking','Country(or dependency)','Population(2020)','Yearly Change','Net Change',
        'Density (P/Km2)','Land Area(Km2)','Migrants(net)','Fertility Rate','Median Age','Urban Pop %','World Share'])
```

### 3) Once the table is scraped, perform some basic data cleaning steps

```python
df.info()#It shows the data types of all columns 
```

#### I observed that some of the columns is having commas,% and cleaning the  column using replace function.

```python
for i in range(len(df['Population(2020)'])):
    df['Population(2020)'][i]=df['Population(2020)'][i].replace(',','')
```

```python
for i in range(len(df['Net Change'])):
    df['Net Change'][i]=df['Net Change'][i].replace(',','')
```

```python
for i in range(len(df['Yearly Change'])):
    df['Yearly Change'][i]=df['Yearly Change'][i].replace('%','')
```

```python
for i in range(len(df['Urban Pop %'])):
    df['Urban Pop %'][i]=df['Urban Pop %'][i].replace('%','')
```

```python
for i in range(len(df['World Share'])):
    df['World Share'][i]=df['World Share'][i].replace('%','')
```

```python
for i in range(len(df['Density (P/Km2)'])):
    df['Density (P/Km2)'][i]=df['Density (P/Km2)'][i].replace(',','')
```

```python
for i in range(len(df['Land Area(Km2)'])):
    df['Land Area(Km2)'][i]=df['Land Area(Km2)'][i].replace(',','')
```

```python
for i in range(len(df['Migrants(net)'])):
    df['Migrants(net)'][i]=df['Migrants(net)'][i].replace(',','')
```

### 9) Create appropriate graphs to examine the distribution of Population, median age and  Urban population and Yearly change.

```python
df_2020_population=df[:50]
```

```python
df_2020_population
```

### The Bar Graph Shows the Top 50 countries Population

```python
plt.figure(figsize=(20,7))
sns.barplot(x = df_2020_population['Country(or dependency)'], y = df_2020_population['Population(2020)'], palette = 'dark')
plt.title('Population of Top 20 countries in 2020',fontsize=15)
plt.xlabel('Countries',fontsize=10)
plt.xticks(rotation=45)
plt.ylabel('Population in Billion',fontsize=10)
plt.show()
```

### The Bar Graph Shows the Top 50 countries Population Median Age.

```python
plt.figure(figsize=(8,25))
sns.barplot(y = df_2020_population['Country(or dependency)'], x= df_2020_population['Median Age'])
plt.title('Population of Top 50 countries in 2020',fontsize=15)
plt.xlabel('Median Age')
plt.ylabel('Country(or dependency)')
plt.yticks(rotation=45)
plt.show()
```

### The Bar Graph Shows the Top 50 countries  Urban Population

```python
plt.figure(figsize=(35,10))
sns.barplot(x = df_2020_population['Country(or dependency)'], y = df_2020_population['Urban Pop %'], palette = 'dark')
plt.title('Population of Top 50 countries in 2020',fontsize=25)
plt.xlabel('Countries',fontsize=25)
plt.ylabel('Urban Pop %',fontsize=25)
plt.xticks(rotation=45)
plt.show()
```

### The Bar Graph Shows the Top 50 countries Yearly Change

```python
plt.figure(figsize=(20,30))
sns.barplot(y = df_2020_population['Country(or dependency)'], x = df_2020_population['Yearly Change'],palette="dark")
plt.title('Population of Top 20 countries in 2020')
plt.xlabel('Yearly Change')
plt.ylabel('Countries')
plt.yticks(rotation=45)
plt.show()
```
